# Parsing datetime data

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/landslides.csv")
df.head()
```

## Check and handle missing values

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df[~df['date'].isnull()]
```

```python
df
```

```python
df['time'].value_counts()
```

```python
df['time'].isna().sum()
```

```python
df['time'] = df['time'].fillna("Not Known")
df
```

```python
df['time'].isna().sum()
```

```python
mean = df['fatalities'].mean()
mean
```

```python
df['fatalities'] = df['fatalities'].fillna(mean)
df
```

```python
df['fatalities'].isna().sum()
```

## Date time parsing using .to_datetime()

```python
df['date'] = pd.to_datetime(df.date)
```

```python
df
```

```python
df['date'].dt.month.plot()
```

```python
new = df['date'].dt.month
```

```python
new
```

```python
new.hist(column="df['date'].dt.month")
```

```python
import seaborn as sns
new = new.dropna()
```
